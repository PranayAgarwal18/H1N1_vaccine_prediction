# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logreg=LogisticRegression()
```

```python
X=vaccinetraindf
y=vaccine_labels.h1n1_vaccine
```

```python
logregmodel=logreg.fit(X,y)
```

```python
print(classification_report(y,logregpredict))
```
